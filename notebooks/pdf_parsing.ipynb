```python
import pymupdf as fitz

pdf_patyh = "/Users/arjunv/Desktop/projects/career-copilot/data/Arjun_V-DS-ML-V1.pdf"
```

```python
document = fitz.open(pdf_patyh)
for page in document:
    text = page.get_text()
    print(text)
```

```python
tables = []
for page in document:
    blocks = page.get_text("blocks")
    for block in blocks:
        if block[6] == 5:  # Check if the block is a table
            tables.append(block)    
print(f"Found {len(tables)} tables in the PDF.")    
```

```python
images = []
for page in document:
    blocks = page.get_text("blocks")
    for block in blocks:
        if block[6] == 1:  # Check if the block is an image
            images.append(block)
print(f"Found {len(images)} images in the PDF.")
```

```python
urls = []
for page in document:
    for link in page.get_links():
        uri = link.get("uri")
        if uri:
            urls.append(uri)

print(f"Found {len(urls)} URLs in the PDF.")
print(urls)
```
